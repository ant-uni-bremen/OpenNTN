After importing all necessary libraries, we now generate a simple NTN channel. The available models are characterized by their scenario (DenseUrban, Urban, or SubUrban), frequency band (S-Band or Ka-Band), direction (uplink or downlink), and their elevation agnle (between 10.0 and 90.0 degrees). We choose the following setup:

```python
scenario = "dur" #dur is DenseUrban
carrier_frequency = 2.0e9 #UL S-Band
direction = "uplink"
elevation_angle = 80.0
```

Next, we setup our antennas. For the satellites, the new antenna type arperture is available, but already existing types such as the VSAT antenna still work. We choose:

```python
ut_array = Antenna(polarization="single",
                    polarization_type="V",
                    antenna_pattern="omni",
                    carrier_frequency=carrier_frequency)

# The satellite will be the base station in our case
bs_array = AntennaArray(num_rows=1,
                        num_cols=4,
                        polarization="dual",
                        polarization_type="VH",
                        antenna_pattern="38.901",
                        carrier_frequency=carrier_frequency)

#Matching the scenario
channel_model = DenseUrban(carrier_frequency=carrier_frequency,
                        ut_array=ut_array,
                        bs_array=bs_array,
                        direction=direction,
                        elevation_angle=elevation_angle)
```

As the NTN channels are made to be used like the existing 38.901 channels, we also need a channel topology. Just note, that the function to generate this topology is different from the one in 38.901. Besides the existing elements, we choose the following addtional parameters:

```python
batch_size = 32
num_ut = 4
sat_height = 600000.0 #the topology takes the satellite height in meters as the base station height

# Generate the topology
topology = gen_ntn_topology(batch_size=batch_size, num_ut=num_ut, scenario=scenario,bs_height=sat_height)

# Set the topology
channel_model.set_topology(*topology)
```

Let's visualize the topology to find first key differences.

```python
channel_model.show_topology()
```

First, it seems as if all UTs are on the same spot. However, this is just caused by all users being in the same cell and an average distance between the users which is very small compared to the satellites height. Next, we see that the users all point directly towards the satellite whereas the satellite points directly to the ground. This is the base setup for all topologies. 

Using this model, we can now generate channel impulse responses and use them as desired to investigate them or use them to simulate transmissions.

```python
num_time_steps = 50
sampling_frequency = 10**(-6)

# path_coefficients [batch size, num_rx, num_rx_ant, num_tx, num_tx_ant, num_paths, num_time_steps]
# path_delays [batch size, num_rx, num_tx, num_paths]
path_coefficients, path_delays = channel_model(num_time_steps, sampling_frequency)
```

However, we can also use those to directly generate the existing OFDM channel implementation in Sionna, which we will do in the next cells. This however also requires the definition of an OFDM resource grid. We choose:

```python
num_streams_per_tx = 1 #as in UL all UTs have one antenna

rg = ResourceGrid(num_ofdm_symbols=14,
                    fft_size=128,
                    subcarrier_spacing=30e3,
                    num_tx=num_ut,
                    num_streams_per_tx=num_streams_per_tx,
                    cyclic_prefix_length=20,
                    pilot_pattern="kronecker",
                    pilot_ofdm_symbol_indices=[2,11])

frequencies = subcarrier_frequencies(rg.fft_size, rg.subcarrier_spacing)

cir = channel_model(1, 1)

# Compute the frequency response
h = cir_to_ofdm_channel(frequencies, *cir, normalize=True)
h = tf.squeeze(h)
```

We can now investigate this frequency response, for example by visualizing its influence on the different subcarriers.

```python
plt.plot(np.abs(h[0,0,0,:]))
#plt.plot(np.imag(h[0,0,0,:]))
plt.legend(["Channel Gain"])
#plt.legend(["Real Part", "Imag Part"])
plt.xlabel("Subcarrier Index")
plt.title("Channel frequency response")
```

The channel responses vary heavily for different elevation angles and scenarios. To showcase that, we will have a look at the channel conidtion number.

```python
def cond_hist(scenario, elevation_angle, sat_height, doppler_enabled=False, num_ut=4, batch_size=1024, direction="uplink"):
    """Generates a histogram of the channel condition numbers"""
    if direction == "uplink":
        carrier_frequency = 2.0e9 #UL S-Band
    else:
        carrier_frequency = 2.2e9 #DL S-Band
        
    ut_array = Antenna(polarization="single",
                        polarization_type="V",
                        antenna_pattern="omni",
                        carrier_frequency=carrier_frequency)

    # The satellite will be the base station in our case
    bs_array = AntennaArray(num_rows=1,
                            num_cols=4,
                            polarization="dual",
                            polarization_type="VH",
                            antenna_pattern="38.901",
                            carrier_frequency=carrier_frequency)
    
    # Setup a CIR generator
    if scenario == "dur":
        channel_model = DenseUrban(carrier_frequency=carrier_frequency,
                                ut_array=ut_array,
                                bs_array=bs_array,
                                direction=direction,
                                elevation_angle=elevation_angle,
                                enable_pathloss=False,
                                enable_shadow_fading=False,
                                doppler_enabled=doppler_enabled)
    elif scenario == "sur":
        channel_model = SubUrban(carrier_frequency=carrier_frequency,
                                ut_array=ut_array,
                                bs_array=bs_array,
                                direction=direction,
                                elevation_angle=elevation_angle,
                                enable_pathloss=False,
                                enable_shadow_fading=False,
                                doppler_enabled=doppler_enabled)
    elif scenario == "urb":
        channel_model = Urban(carrier_frequency=carrier_frequency,
                                ut_array=ut_array,
                                bs_array=bs_array,
                                direction=direction,
                                elevation_angle=elevation_angle,
                                enable_pathloss=False,
                                enable_shadow_fading=False,
                                doppler_enabled=doppler_enabled)

    topology = gen_ntn_topology(batch_size=batch_size, num_ut=num_ut, scenario=scenario,bs_height=sat_height)

    # Set the topology
    channel_model.set_topology(*topology)

    # Generate random CIR realizations
    # As we nned only a single sample in time, the sampling_frequency
    # does not matter.
    cir = channel_model(1, 1)
    
    # Compute the frequency response
    h = cir_to_ofdm_channel(frequencies, *cir, normalize=True)
    
    h = tf.squeeze(h)
    h = tf.transpose(h, [0,3,1,2])
    
    # Compute condition number
    c = np.reshape(np.linalg.cond(h), [-1])

    # Compute normalized histogram
    hist, bins = np.histogram(c, 100)
    hist = hist/np.sum(hist)
    return bins[:-1], hist
```

```python
for scenario in ["dur", "sur", "urb"]:
    bins, hist = cond_hist(scenario, elevation_angle, sat_height)
    plt.plot(bins, np.cumsum(hist))
    plt.legend(["DUR", "SUR", "URB"])
    plt.xlabel("Channel Condition Number")
    plt.ylabel("CDF")
    plt.title("CDF of the channel condition number")
```

This looks really bad. However, the reason lies in the really large doppler influence, which can cause full phase rotations between two OFDM symbols, making the channel very difficult to predict. Thus many research questions might want to assume the doppler to be known or compensated. As a shortcut, we added the possibility to disable it in the channel settings, resulting in the following diagrams.

```python
for scenario in ["dur", "sur", "urb"]:
    bins, hist = cond_hist(scenario, elevation_angle, sat_height, doppler_enabled=False)
    hist = hist/np.sum(hist)
    plt.plot(bins, np.cumsum(hist))
    plt.legend(["DUR", "SUR", "URB"])
    plt.xlabel("Channel Condition Number")
    plt.ylabel("CDF")
    plt.title("CDF of the channel condition number")
```

EVen though the channel looks much more manageble now, it still is much worse compared to the terrestrial one. This comes from the channels between indivudal users being very different, so by reducing that, we can get

```python
for scenario in ["dur", "sur", "urb"]:
    bins, hist = cond_hist(scenario, elevation_angle, sat_height, doppler_enabled=True,num_ut=2, batch_size=4096)
    hist = hist/np.sum(hist)
    plt.plot(bins, np.cumsum(hist))
    plt.legend(["DUR", "SUR", "URB"])
    plt.xlabel("Channel Condition Number")
    plt.ylabel("CDF")
    plt.xlim(0,100)
    plt.title("CDF of the channel condition number")
```

Lastly, we can still adjust the other factors, such as the elevation angle for more and less managable results 

```python
for scenario in ["dur", "sur", "urb"]:
    bins, hist = cond_hist(scenario, elevation_angle=80.0, sat_height=600000.0, doppler_enabled=False,num_ut=2)
    hist = hist/np.sum(hist)
    plt.plot(bins, np.cumsum(hist))
    plt.legend(["DUR", "SUR", "URB"])
    plt.xlabel("Channel Condition Number")
    plt.ylabel("CDF")
    plt.title("CDF of the channel condition number")
```

We see that the channels are very challenging, especially for handling multiple users at once. For the last part, we will have a look at the channel impulse response once again. Thee underlying channel model defines the channel through a number of clusters generated by the transmitter, which in turn generate a number of rays arriving at the receiver. This transmission is defined between each transmitter and receiver element. Each of the rays is defined by its path coefficient (power and phase of the rays as one complex number) and its time of arrival (relative delay compared to the first ray). For better understanding of this, we visualize a few examples.

```python
num_time_steps = 10
sampling_frequency = 10**(5)

# path_coefficients [batch size, num_rx, num_rx_ant, num_tx, num_tx_ant, num_paths, num_time_steps]
# path_delays [batch size, num_rx, num_tx, num_paths]
path_coefficients, path_delays = channel_model(num_time_steps, sampling_frequency)
```

First, let's have a look at the different arrival times for one user

```python
plt.plot(path_delays[0,0,0,1:])
plt.yscale('log')
plt.xlabel("Path Index")
plt.ylabel("Delay in s")
plt.title("Normalized Delays of All Paths for one Tx Rx Pair")
```

```python
plt.plot(tf.reduce_mean(tf.abs(path_coefficients[0,0,0,0,0,:,:]),axis=-1))
plt.yscale('log')
plt.xlabel("Path Index")
plt.ylabel("Path Gain")
plt.title("Gain of All Paths for one Tx Rx Pair")
```

As we see, the delays are grouped, with the first being 0, followed by six close ones and a very large one. Accordingly, we can have a look at the power of each ray, which is dropping correspondingly.
Lastly, let's have a look at the phase rotation of the model.

```python
for idx in range(5):
    plt.scatter(tf.math.real(path_coefficients[0,0,0,0,0,0,idx]),tf.math.imag(path_coefficients[0,0,0,0,0,0,idx]))
plt.legend(["1","2","3","4","5"])
# Adding x and y axes through the origin
plt.axhline(0, color='black', linewidth=0.5)  # Horizontal axis (y=0)
plt.axvline(0, color='black', linewidth=0.5)  # Vertical axis (x=0)

# Adding labels for the axes
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')

real_part = tf.math.real(path_coefficients[0,0,0,0,0,0,:])
imag_part = tf.math.imag(path_coefficients[0,0,0,0,0,0,:])

max_limit = max(max(abs(real_part)), max(abs(imag_part)))*1.1
plt.xlim(-max_limit, max_limit)
plt.ylim(-max_limit, max_limit)

# Displaying the plot
plt.grid(True)  # Optional: add a grid for better visualization

plt.title("Symbols at Five Points in Time Experiencing Doppler Phase Shift")
```

The above grid shows the influence of the doppler shift on five rays of the same path.
